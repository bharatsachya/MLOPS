```python
x = np.linspace(0,10,100)
y = np.sin(x)
a = np.array([0,30,45,60,90])
b =np.deg2rad(a)
z = np.sin(b)
plt.plot(a,b)
plt.xlabel('x-axo')
plt.ylabel('y-axis')
plt.title('Sine Curve')
plt.show()
```

```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x,y)
plt.xlabel('x-axo')
plt.ylabel('y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
# histogram
data = np.random.normal(100,20,1000)
plt.hist(data,bins=20,color='purple')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
#Box Plot
data = [np.random.normal(50,10,100),np.random.normal(80,20,100),np.random.normal(40,15,100)]
plt.boxplot(data)
plt.xlabel('x-axo')
plt.ylabel('y-axis')
plt.title('Box Plot')
plt.show()
```

## __Structured Array__

```python
dt = np.dtype([('name','S10'),('age','int')])
a = np.array([('John',21),('Smith',23)],dtype=dt)
a
```

```python
person = a[0]

name = person['name']
print(name)

age = person['age']
print(age)
```

```python
avg_age = a['age'].mean()
avg_age

people_older_than_22 = a[a['age']>22]
print(people_older_than_22)
```

```python
data_dtype = np.dtype([('name','S10'),('age','int'),('weight','float')])

a = np.dtype([('height','i4'),('person',data_dtype)])

b = np.array([(163,('Lovanshu',21,86.0)),(100,('Loki',21,50.0))],dtype=a)
print(b)
```

```python
person = b[0]['person']

name = person['name']
name = 'Ganesh'
person['age'] +=1
print(person)
```
